```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
server = open("C:\\Users\\dharu\\OneDrive\\Desktop\\ISROSpaceHackathon\\ssl_access.txt")
con_server = server.readlines()
```

```python
cols = ['ip_address', 'identd', 'user', 'timestamp', 'method', 'url', 'protocol', 'status', 'size', 'referer', 'user_agent']
```

```python
con_server
```

```python
log_entry = '172.26.3.214 - - [06/Nov/2023:00:00:02 +0530] "GET /scan_m/cas_userstats/getUserStats.php?sdate=2023-11-06&edate=2023-11-06 HTTP/1.1" 200 93 "-" "python-requests/2.5.1 CPython/3.8.5 Linux/3.10.0-327.el7.x86_64"'
pattern = re.compile(r'(?P<ip>[\d.]+) - - \[(?P<timestamp>.*?)\] "(?P<method>.*?) (?P<url>.*?) (?P<protocol>.*?)" (?P<status>\d+) (?P<size>\d+) "(?P<referer>.*?)" "(?P<user_agent>.*?)"')
match = pattern.match(log_entry)
if match:
    values = [match.group('ip'), None, None, match.group('timestamp'), match.group('method'), match.group('url'),
              match.group('protocol'), match.group('status'), match.group('size'), match.group('referer'), match.group('user_agent')]
```

```python
data = []
for i in range(len(con_server)):
    match = pattern.match(log_entry)

    # If there is a match, extract the values in the specified order
    if match:
        # Extract values using the group indices from the pattern
        values = [match.group('ip'), None, None, match.group('timestamp'), match.group('method'), match.group('url'),
                  match.group('protocol'), match.group('status'), match.group('size'), match.group('referer'), match.group('user_agent')]
        data.append(values)
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.columns = cols
```
