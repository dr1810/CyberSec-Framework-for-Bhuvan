```python
cnt = firewall["srcip"].value_counts()
unusual_srcip = firewall["srcip"].isin(cnt[cnt<10].index)
print(firewall[unusual_srcip][non]) #Unusual Source IP Address Detection
print("Information")
print(firewall[unusual_srcip]["srcip"].unique())
```

```python
cnt = firewall["dstip"].value_counts()
unusual_srcip = firewall["dstip"].isin(cnt[cnt<10].index)
print(firewall[unusual_srcip][non]) #Unusual Destination IP
print("Information")
print(firewall[unusual_srcip]["dstip"].unique())
```
